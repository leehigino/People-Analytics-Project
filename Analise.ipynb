```python
import pandas as pd
import numpy as np
import re

df = pd.read_csv("HR-Employee-Attrition.csv")
display(df)
display(list(df.columns))
display(df.info())
```

```python
#Cleaning stage

df_clean = df.drop(columns=["EmployeeCount","StandardHours", "Over18"])



display(df_clean)
```

```python
# Transform stage - Feature engineering

# Is Turnover
df_clean["IsTurnover"] = np.where(df_clean["Attrition"]=="Yes", 1, 0)

# Age Band
df_clean["AgeBand"] = pd.cut(
    df_clean["Age"], 
    bins=[18, 30, 40, 65], 
    labels=["<30", "30-40", "40+"]
    )

# Ternure Band
df_clean["TenureBand"] = pd.cut(
    df_clean["YearsAtCompany"],
    bins=[-1, 1, 3, 5, 10, 40],
    labels=["<1", "1-3", "3-5", "5-10", "10+"]
)

# Distance Band
df_clean["DistanceBand"] = pd.cut(
    df_clean["DistanceFromHome"],
    bins=[0, 5, 10, 20, 30],
    labels=["Very Near", "Near", "Medium", "Far"]
    )

# Promotion Gap Band
df_clean["PromotionGapBand"] = pd.cut(
    df_clean["YearsSinceLastPromotion"],
    bins=[-1, 1, 3, 5, 10, 15],
    labels=["Recent", "1-3", "3-5", "5-10", "10+"]
)

# Standard Satisfaction Scale (1–4)
satisfaction_map = {
    1: "Low",
    2: "Medium",
    3: "High",
    4: "Very High"
}

df_clean["JobSatisfactionLabel"] = df_clean["JobSatisfaction"].map(satisfaction_map)
df_clean["EnvironmentSatisfactionLabel"] = df_clean["EnvironmentSatisfaction"].map(satisfaction_map)
df_clean["RelationshipSatisfactionLabel"] = df_clean["RelationshipSatisfaction"].map(satisfaction_map)

# Worklife Balance Scale
worklife_map = {
    1: "Poor",
    2: "Fair",
    3: "Good",
    4: "Excellent"
}

df_clean["WorklifeBalanceLabel"] = df_clean["WorkLifeBalance"].map(worklife_map)

# Job Satisfaction Risk (1–4)
# Lower satisfaction = higher risk

df_clean["JobSatisfactionRisk"] = 1 - (df_clean["JobSatisfaction"] / 4)

# Work-Life Balance Risk (1–4)
# Poor balance = higher risk

df_clean["WorklifeRisk"] = 1 - (df_clean["WorkLifeBalance"] / 4)


# Tenure Risk (Years at Company)
# Shorter tenure = higher risk

df_clean["TenureRisk"] = np.where(
    df_clean["YearsAtCompany"] < 2, 1,
    np.where(df_clean["YearsAtCompany"] < 5, 0.5, 0.2)
)

# Overtime Risk
# Overtime = higher risk

df_clean["OvertimeRisk"] = np.where(
    df_clean["OverTime"] == "Yes", 1, 0
)

# Distance from Home Risk
# Longer distance = higher risk

df_clean["DistanceRisk"] = np.where(
    df_clean["DistanceFromHome"] > 20, 1,
    np.where(df_clean["DistanceFromHome"] > 10, 0.5, 0.2)
)

# Promotion Gap Risk
# Long time without promotion = higher risk

df_clean["PromotionGapRisk"] = np.where(
    df_clean["YearsSinceLastPromotion"] >= 5, 1,
    np.where(df_clean["YearsSinceLastPromotion"] >= 2, 0.5, 0.2)
)

# Job Level Risk
# Lower job level = higher risk

df_clean["JobLevelRisk"] = 1 - (
    df_clean["JobLevel"] / df_clean["JobLevel"].max()
)


# TURNOVER RISK SCORE (0–100)
# Weighted composite score

df_clean["TurnoverRiskScore"] = (
    df_clean["JobSatisfactionRisk"] * 0.25 +
    df_clean["WorklifeRisk"] * 0.20 +
    df_clean["TenureRisk"] * 0.15 +
    df_clean["OvertimeRisk"] * 0.15 +
    df_clean["DistanceRisk"] * 0.10 +
    df_clean["PromotionGapRisk"] * 0.10 +
    df_clean["JobLevelRisk"] * 0.05
) * 100


# TURNOVER RISK LEVEL (BUCKETS)

df_clean["TurnoverRiskLevel"] = pd.cut(
    df_clean["TurnoverRiskScore"],
    bins=[0, 30, 60, 100],
    labels=["Low Risk", "Medium Risk", "High Risk"]
)

display(df_clean)

display(list(df_clean.columns))


```

```python
import re

df_clean.columns = [
    re.sub('([a-z0-9])([A-Z])', r'\1_\2', col).lower()
    for col in df_clean.columns
]

display(df_clean.columns.tolist())
display(df_clean.head())
```
